```python
data_x = np.array([1,2,3])
data_y= np.array([1,2,3])
learning_rate = 0.01
m_guess=0
c_guess=0
# y_guess=m_guess*data_x+c_guess
# loss_function=(data_y-y_guess)**2

def linear_equation(m,x,c):
    return m*x+c

def loss_function(x,y,m,c):
    return (y-m*x-c)

def slope_loss_derivative(x,y,m,c):
    slope_step=-2*loss_function(x,y,m,c)*x
    return slope_step

def intercept_loss_derivative(x,y,m,c):
    intercept_step=2*loss_function(x,y,m,c)
    return intercept_step

    
def gradient_descent(learning_rate,array_x,array_y,m,c):
    slopeStep=0
    interceptStep=0
    for i in range(len(array_x)):
        individual_slope_derv=slope_loss_derivative(array_x[i],array_y[i],m,c)
        slopeStep+=individual_slope_derv
        
        individual_intercept_derv=intercept_loss_derivative(array_x[i],array_y[i],m,c)
        interceptStep+=individual_intercept_derv
        
    print('step_slope:',slopeStep)
    print('step_intercept',interceptStep)
    if (np.abs(slopeStep)>0.05 or np.abs(interceptStep)>0.05):
        if np.abs(slopeStep)<np.abs(interceptStep):
            c_guess=c-interceptStep*learning_rate
            m_guess=m
        elif np.abs(interceptStep)<np.abs(slopeStep):
            m_guess=m-slopeStep*learning_rate
            c_guess=c
        print('*m_guess', m_guess)
        print('*c_guess',c_guess)
        return gradient_descent(0.001,data_x,data_y,m_guess,c_guess)
    else:
        result=[m,c]
        print("hello")
        
    return result


```
